```python
import gymnasium as gym
import numpy as np
import sys
import pygame
```

```python
class starEnv(gym.Env):
    def __init__(self, grid_size=10):
        
        super(starEnv, self).__init__()
        self.grid_size=grid_size
        self.cell_size = 100
        self.state = None
        self.reward = 0
        self.info = {}
        self.goal = np.array([9,9])
        self.done =False
        self.hurdle_states = []

        self.action_space = gym.spaces.Discrete(4)

        self.observation_space = gym.spaces.Box(low=0, high=grid_size-1, shape=(2,), dtype=np.int32)

        pygame.init()

        self.screen = pygame.display.set_mode((self.cell_size*self.grid_size, self.cell_size*self.grid_size))

    def reset(self):
        self.state = np.array([0,0])
        self.done = False
        self.reward = 0

        self.info["Distance to goal"] = np.sqrt(
        (self.state[0]-self.goal[0])**2 + 
        (self.state[1]-self.goal[1])**2
        )

        return self.state, self.info

    def add_hurdle_states(self, hurdle_state_coordinates):
        self.hurdle_states.append(np.array(hurdle_state_coordinates))

    def step(self,action):

        if action == 0 and self.state[0] > 0:
            self.state[0] -= 1

        if action == 1 and self.state[0] < self.grid_size - 1:
            self.state[0] += 1

        if action == 2 and self.state[1] < self.grid_size - 1:
            self.state[1] += 1

        if action == 3 and self.state[1] < self.grid_size - 1:
            self.state[1] -= 1
        print(self.state,"Goal:",self.goal)
        if np.equal(self.state,self.goal):
            self.reward += 10
            self.done = True

        elif True in [np.array_equal(self.state,each_hurdle) for each_hurdle in self.hurdle_states]:
            self.reward += -2
            self.done = True

        else:
            self.reward += -0.05
            self.done = False

        self.info["Distance to goal"] = np.sqrt(
        (self.state[0]-self.goal[0])**2 + 
        (self.state[1]-self.goal[1])**2
        )

        return self.state, self.reward, self.done, self.info

        

    def render(self):
        for event in pygame.event.get():
            if event == pygame.QUIT:
                pygame.quit()
                sys.exit()

        self.screen.fill((255,255,255))

        for y in range(self.grid_size):
            for x in range(self.grid_size):
                grid = pygame.Rect(y*self.cell_size, x*self.cell_size, self.cell_size, self.cell_size)
                pygame.draw.rect(self.screen, (255,255,255), grid, 1)

        # Draw the Goal-state:
        goal = pygame.Rect(self.goal[1]*self.cell_size, self.goal[0]*self.cell_size, self.cell_size, self.cell_size)
        pygame.draw.rect(self.screen, (0,255,0), goal)

        # Draw the hell-states:
        for each_hell in self.hell_states:
            hell = pygame.Rect(each_hell[1]*self.cell_size, each_hell[0]*self.cell_size, self.cell_size, self.cell_size)
            pygame.draw.rect(self.screen, (255,0,0), hell)

        # Draw the agent:
        agent = pygame.Rect(self.state[1]*self.cell_size, self.state[0]*self.cell_size, self.cell_size, self.cell_size)
        pygame.draw.rect(self.screen, (0,0,0), agent)

        # Update contents on the window:
        pygame.display.flip()

    def close(self):
        pygame.display.flip()
        print("Window close: ",pygame.WINDOWCLOSE)
        pygame.quit()

        
                

```

```python
star_env = starEnv(grid_size=10)

star_env.add_hurdle_states(hurdle_state_coordinates=(2,5))
star_env.add_hurdle_states(hurdle_state_coordinates=(2,6))
star_env.add_hurdle_states(hurdle_state_coordinates=(2,7))
star_env.add_hurdle_states(hurdle_state_coordinates=(2,8))
```

```python
observation,info = star_env.reset()

print(f"Initial state: {observation}, Info: {info}")

# clock = pygame.time.Clock()
# FPS = 30
back = False
for _ in range(30):
    action = int(input("Choose the war: "))

    new_state, reward, done, info = star_env.step(action)
    print(f"New state: {new_state}, Reward: {reward}, Done: {done}, Info: {info}")
    # pygame.display.update()

    # Check for termination condition
    if done:
        # pygame.event.pump()
        back=True
        sys.exit()
        break
        
     # Render the environment
    # if back==True:
    #     star_env.close()
    # else:
    #     star_env.render()
```
